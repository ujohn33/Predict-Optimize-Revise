```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scienceplots
```

```python
ES_1 = pd.read_csv('results/energy_score_phase1.csv', index_col=0)
ES_3 = pd.read_csv('results/energy_score_phase3.csv', index_col=0)
ES_1
```

```python
# set style to seaborn
plt.style.use(['science','ieee'])
# proportion 2:1
fig, ax = plt.subplots(1, 1, figsize=(15, 5))
ax.plot(ES_1.index, ES_1['ES'], label='Phase 1', marker='^', linestyle='dotted', 
        linewidth=5, markersize=15, color='#333333')
ax.grid()
# make x ticks bigger
ax.tick_params(axis='x', labelsize=18)
ax.tick_params(axis='y', labelsize=18)
ax.set_xlabel('Forecast revision period', fontsize=20)
ax.set_ylabel('Energy Score', fontsize=20)
plt.show()
```

```python
# set style to seaborn
plt.style.use(['science','ieee'])
# proportion 2:1
fig, ax = plt.subplots(1, 1, figsize=(15, 5))
ax.plot(ES_3.index, ES_3['ES'], label='Phase 1', marker='^', linestyle='dotted', 
        linewidth=5, markersize=15, color='#333333')
ax.grid()
# make x ticks bigger
ax.tick_params(axis='x', labelsize=18)
ax.tick_params(axis='y', labelsize=18)
ax.set_xlabel('Forecast revision period', fontsize=20)
ax.set_ylabel('Energy Score', fontsize=20)
plt.show()
```
