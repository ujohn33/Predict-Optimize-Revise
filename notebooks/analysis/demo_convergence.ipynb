```python
import os, inspect
import sys
os.chdir(os.path.dirname(os.getcwd()))
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import scienceplots

plt.style.use('science')

# Define the function for competitive difference bound
def competitive_difference(v, A, B, a):
    return A / v + B * np.sqrt(1 - a**(2 * (v + 1)))

# Define commitment levels
v = np.linspace(1, 30, 300)

# Case 2: Deterministic
A2, B2, a2 = 5, 6, 0.99
cd_deterministic = competitive_difference(v, A2, B2, a2)

# Case 2: Stochastic (lower B to reflect reduced forecast penalty)
A2_s, B2_s, a2_s = A2, 4, a2
cd_stochastic = competitive_difference(v, A2_s, B2_s, a2_s)

plt.figure(figsize=(10, 6))
plt.plot(v, cd_deterministic, label=f'Deterministic (A={A2}, B={B2}, a={a2})', color='darkblue', linewidth=2)
plt.plot(v, cd_stochastic, label=f'Stochastic (A={A2_s}, B={B2_s}, a={a2_s})', color='darkblue', linestyle='--', linewidth=2)

plt.xlabel('Commitment level $v$', fontsize=16)
plt.ylabel('Expected Competitive Difference', fontsize=16)

# make x axis logarithmic
plt.xscale('log')
plt.legend(fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True, alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Define the range of v
v = np.arange(1, 50)

# Parameters for both cases
a = 0.99
B = 6

# Case 1: Low switching cost
A1 = 0.2
cost1 = A1 / v + B * (1 - a ** (2 * (v + 1)))

# Case 2: High switching cost
A2 = 5
cost2 = A2 / v + B * (1 - a ** (2 * (v + 1)))

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(v, cost1, label=r'Low switching costs: $A={}, B={}, a={}$'.format(A1, B, a), color='royalblue', linewidth=2)
plt.plot(v, cost2, label=r'High switching costs: $A={}, B={}, a={}$'.format(A2, B, a), color='navy', linewidth=2)

plt.xlabel("Commitment level $v$", fontsize=16)
plt.ylabel("Expected competitive difference", fontsize=16)

# make x axis logarithmic
plt.xscale('log')
#plt.title("Trade-off in FHC: Switching Cost vs Forecast Error", fontsize=16)
plt.legend(fontsize=16)
plt.legend(fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True, alpha=0.7)
plt.tight_layout()
plt.show()
```
