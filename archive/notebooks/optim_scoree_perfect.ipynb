```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scienceplots
```

```python
df_perfect = pd.read_csv('results/scores_perfectMPC_phase3.csv', index_col=0)
# keep only rows and columns from 1 to 12
df_perfect = df_perfect.iloc[:12, :12]
df_perfect
```

```python
plt.style.available
```

```python
# get a diagonal line
plt.style.use(['seaborn-whitegrid', 'science', 'ieee'])
diag = np.diag(df_perfect)
fig, ax = plt.subplots(1, 1, figsize=(15, 5))
# plot the diagonal vs index of df
ax.plot(df_perfect.index, diag, marker='*', linestyle='dotted', 
        linewidth=5, markersize=20, color='blue', markeredgecolor='#333333',  markerfacecolor='#333333')
ax.tick_params(axis='x', labelsize=14)
ax.tick_params(axis='y', labelsize=14)
plt.xlabel('Optimiztion Revision period', fontsize=18)
plt.ylabel('KPI score',  fontsize=18)
# add title
#plt.title('Optimization scores on the Diagonal', fontsize=18)
plt.show()
```

```python
df = pd.read_csv('results/opt_and_forecast_revision_study_12x12_phase3.csv')
df = df[df['phase'] == 3]
# make a pivot table from steps_skip_optimizaiton and steps_skip_forecast and tc as values
df_pivot = df.pivot_table(index='steps_skip_optimization', columns='steps_skip_forecast', values='tc')
# in df_pivot make all cells where column index is smaller than row index NaN
for i in range(0,12):
    for j in range(0,12):
        if j < i:
            df_pivot.iloc[i,j] = np.nan
df_pivot
```

```python
# get a diagonal line
sns.set_style("darkgrid")
diag = np.diag(df_pivot.values / df_perfect.values[0][0])
fig, ax = plt.subplots(1, 1, figsize=(15, 5))
# plot the diagonal vs index of df
ax.plot(df_perfect.index, diag, marker='*', linestyle='dotted', 
        linewidth=5, markersize=20, color='blue', markeredgecolor='#333333',  markerfacecolor='#333333')
ax.tick_params(axis='x', labelsize=14)
ax.tick_params(axis='y', labelsize=14)
plt.xlabel('Optimiztion Revision period', fontsize=18)
plt.ylabel('Regret',  fontsize=18)
# add title
#plt.title('Optimization scores on the Diagonal', fontsize=18)
plt.show()
```
